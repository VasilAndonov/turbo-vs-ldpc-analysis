# Comparative Analysis of Turbo and LDPC Codes

## Abstract

Turbo Codes and Low-Density Parity-Check (LDPC) Codes are advanced forward error correction (FEC) techniques used in modern digital communication systems to improve reliability over noisy channels. This project performs a comparative analysis of Turbo and LDPC codes, focusing on Bit Error Rate (BER), decoding complexity, and convergence behavior under controlled conditions. Both codes are simulated over an Additive White Gaussian Noise (AWGN) channel with Binary Phase Shift Keying (BPSK) modulation. Metrics such as BER, decoding time, iteration count, and code rate are evaluated. The study aims to determine under which conditions each coding scheme outperforms the other, providing insights into their practical trade-offs and application in 4G and 5G systems.

## Introduction

Error correction is a critical requirement in communication systems to maintain data integrity in the presence of noise. Noise, interference, and channel impairments can cause bit errors, reducing reliability. Forward error correction (FEC) techniques, such as Turbo and LDPC codes, improve system performance without requiring retransmission.

Turbo codes were introduced in the 1990s and became standard in 4G LTE systems due to their near-Shannon-limit performance with moderate complexity. LDPC codes, originally proposed by Gallager in the 1960s and revived for modern communications, are widely adopted in 5G standards for their high performance and efficient parallelizable decoding.

The motivation for this study is to understand and quantify the differences between these two prominent coding schemes. Key research questions include:

- Under which conditions does Turbo outperform LDPC?  
- Under which conditions does LDPC outperform Turbo?  
- How do decoding complexity and convergence differ?  
- How does code rate affect performance?
